```python
%matplotlib widget
import mpl_toolkits.mplot3d as a3
import matplotlib.colors as colors
import pylab as plt
import numpy as np

camsize=1
_l = 3*camsize
_w=1*camsize
cam = np.array([[[0,0,0], [_w, -_w, _l], [_w, _w, _l]],
                [[0,0,0], [_w, _w, _l], [-_w, _w, _l]],
                [[0,0,0], [-_w, _w, _l], [-_w, -_w, _l]],
                [[0,0,0], [-_w, -_w, _l], [_w, -_w, _l]]])
ax = plt.figure().add_subplot(projection='3d')

vtx = cam
tri = a3.art3d.Poly3DCollection(vtx)
tri.set_color('tab:blue')
tri.set_edgecolor('k')
ax.add_collection3d(tri)
ax.set_xlim([-5, 5])
ax.set_ylim([-5, 5])
ax.set_zlim([-5, 5])

plt.show()
```
